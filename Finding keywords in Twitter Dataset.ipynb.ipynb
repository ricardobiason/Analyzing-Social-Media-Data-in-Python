Counting known keywords is one of the first ways that we can analyze text data in a Twitter dataset. In this dataset, we're going to count the number of times specific hashtags occur in a collection of tweets about data science. To this end, ywe're going to use the string methods in the pandas Series object to do this.

Counting words is the most basic step we can take in automating the analysis of text. It allows us to convert words into numbers. In practice, counting words can tell us how many times a company, product, or hashtag is mentioned.

```python
import pandas as pd
import numpy as np
```

```python
# Import Data
Dataset = pd.read_csv(r"C:\Users\Ricardo S. Biason\OneDrive\Ric\DataCamp\Marketing Analytics for Python\Analyzing Social Media Data in Python\Data Science Hashtag dataset.csv")
```

```python
Dataset.head(n=5)
```

```python
# Flatten the tweets and store them as Flat_tweets

flat_tweets = Dataset
```

```python
# # Convert to DataFrame

ds_tweets = pd.DataFrame(flat_tweets)
```

```python
# Find mentions of #python in 'text'

python = ds_tweets['text'].str.contains('#python', case = False)
```

```python
# Print proportion of tweets mentioning #python

print("Proportion of #python tweets:", np.sum(python) / ds_tweets.shape[0])
```
